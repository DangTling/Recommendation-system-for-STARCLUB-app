```python
selected_columns = ['id', 'title', 'artist_name', 'category_name', 'description', 'lyrics']
songs_list = [{key: song[key] for key in selected_columns} for song in songs]

# songs_list = filtered_songs.to_dict(orient='records')

# songs_list
print(songs_list)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

# combinedSongValue = [
#     f"Title: {song['title']}, Artist: {song['artist_name']}, Category: {song['category_name']}, "
#     f"Description: {song['description']}"
#     for song in songs_list
# ]

# songs_embeddings = model.encode(combinedSongValue)

songs_embeddings = []
for song in songs_list:
    text_combined = song['title'] + ' ' + song['artist_name']
    text_vector = model.encode(text_combined)
    author_vector = model.encode(song['category_name'])
    category_vector = model.encode(song['description'])
    combined_vector = np.hstack([text_vector, author_vector, category_vector])
    songs_embeddings.append(combined_vector)
```

```python
client = QdrantClient(host="localhost", port=6333)
client.recreate_collection(
    collection_name="songs_collection_database",
    vectors_config=models.VectorParams(
        size=384*3, 
        distance=models.Distance.COSINE 
    )
)
```

```python
for idx, (song, embedding) in enumerate(zip(songs_list, songs_embeddings)):
    client.upsert(
        collection_name='songs_collection_database',
        points=[
            {
                'id': int(song['id']),
                'vector': embedding.tolist(),
                'payload': song  
            }
        ]
    )
```

```python
def get_recommendations(song_title, song_artist, song_category, song_description, top_k=10):
    # query_description = f"{song_title} - {song_artist} - {song_category} - {song_description}"
    # embedding = model.encode([query_description])[0]

    title_artist_vector = model.encode(song_title + " " + song_artist)
    category_vector = model.encode(song_category)
    description_vector = model.encode(song_description)

    query_vector = np.hstack([title_artist_vector, category_vector, description_vector])
    
    search_result = client.search(
        collection_name='songs_collection_database',
        query_vector=query_vector.tolist(),
        limit=top_k
    )
    
    recommendations = [res.payload for res in search_result]
    return recommendations

recommendations = get_recommendations(
    song_title="Có Lẽ Quá Khó Để Quên Một Người",
    song_artist="1nG",
    song_category="Nhạc trẻ",
    song_description="Bài hát Có Lẽ Quá Khó Để Quên 1 Người - Nam Lee & 1nG thuộc Album Có Lẽ Quá Khó Để Quên 1 Người phát hành năm 2019"
)
print(recommendations)
```

```python
def print_recommendations(recommendations):
    for i, song in enumerate(recommendations, start=1):
        print(f"{i}. Title: {song['title']}")
        print(f"   Artist: {song['artist_name']}")
        print(f"   Category: {song['category_name']}")
        print(f"   Description: {song['description']}")
        print("-" * 40) 

print_recommendations(recommendations)
```
